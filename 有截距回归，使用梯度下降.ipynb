```python
#如果有截距，不能求导，倒数最小求解问题
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda x : (x-4.3)**2 + 7.5*x + 6
```

```python
x = np.linspace(-5,9,100)
plt.plot(x,f(x))
```

```python
#求导
g = lambda x: 2*x -1.1
#要让g = 0 ,x = 0.55 所以0.55有最小值
```

```python
#
result = []
#一开始蒙一个数,设置的起始值假设最小
v_min = np.random.randint(-10,10,size =1)[0]

#该值记录梯度下降过程中，上一步的值
v_min_last = v_min + 1

#什么时候梯度下降结束，精度的退出条件
precision = 0.0001

#再一个退出条件，循环次数

max_time = 3000
count = 0
print("-----------------s随机生成的最小值", v_min)
#步幅
step = 0.01

while True:

    if np.abs(v_min - v_min_last) < precision:
        break
    if count>max_time:
        break
    else:
        v_min_last = v_min

        v_min = v_min -g(v_min)*step
        result.append(v_min)
        print('----------------------梯度下降更新的最小值 ',v_min)
        count+=1
```

```python
x = np.linspace(0,10,100)
plt.figure(figsize=(9,9))
plt.plot(x,f(x))
plt.plot(x,g(x))
#绘制最小值的点result
plt.scatter(result,f(np.array(result)),marker="*",c="r")
```
