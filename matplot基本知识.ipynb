```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#只包含单一曲线的图
x = np.linspace(0,2*np.pi)
y = np.sin(x)
plt.plot(x,y)
```

```python
#两条线
y2 = np.cos(x)
plt.plot(x,y)
plt.plot(x,y2)
```

```python
plt.plot(x,y,x,y2,marker = 'o')
```

```python
#网格线
plt.grid()
```

```python
#子图
plt.figure(figsize=(12,3))
ax = plt.subplot(1,3,1)
ax.plot(x,y)
ax.grid()

ax = plt.subplot(1,3,2)
ax.plot(x,y)
ax.grid(c='r', ls=':', lw=3,alpha=0.2)

ax = plt.subplot(1,3,3)
ax.plot(x,y)
ax.grid()
```

```python
#坐标轴上下限
plt.plot(x,y)
```

```python
#利用坐标轴标签
#plt.axix off(关闭) equal(间隔相等)
plt.plot(x,y)
plt.xlim(-1.5,1.5)
plt.xlabel('nima',fontsize = 20,color = 'red')
plt.title("caonima")
```

```python
#绘制图例
y2 = np.cos(x)
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(['sin','cos'],ncol=2)
```
