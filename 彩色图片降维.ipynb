```python
#图片灰度处理
#根本上来说是最后一维变没了
import matplotlib.pyplot as plt
import numpy as np
```

```python
tupian = plt.imread('./cat.jpg')
```

```python
plt.imshow(tupian)
```

```python
gray = tupian.min(axis = -1)
plt.imshow(gray,cmap=plt.cm.gray)
```

```python
gray = tupian.mean(axis = 2)
plt.imshow(gray,cmap=plt.cm.gray)
```

```python
w = np.array([0.299,0.587,0.114])
gray = tupian.dot(w)
plt.imshow(gray,cmap=plt.cm.gray)
```
