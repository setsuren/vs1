```python
##用pyplot画图
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s = Series(np.random.randint(0,10,size=10))
s
s.plot(kind='pie')
```

```python
df = DataFrame(np.random.randint(0,150,size=(10,2)),index=list('ABCDEFGHIJ'),columns=['Python','En'])
df['Math'] = df['Python'].map(lambda x : x + np.random.randint(-10,10))
df
```

```python
df.plot(kind='bar')
```

```python
s = Series(np.random.randn(10000))
```

```python

s.plot(kind='hist',bins = 1000, density = True)

s.plot(kind ='kde',c = 'red')
```

```python
df.plot(x = 'Python', y = 'Math',kind = 'scatter')
```

```python
plt.figure(figsize=(12,9))
ax=plt.subplot(1,1,1)
_ = pd.plotting.scatter_matrix(df,ax=ax)
```
