```python
#插值
#空数据的补全
#线性插值：interp
#拉格朗日插值法
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.interpolate import lagrange
from scipy import interp
```

```python
x = np.linspace(0,2*np.pi,10)
y = np.sin(x)
plt.plot(x,y)
```

```python
X = np.linspace(0,2*np.pi,40)
#线性插值
Y = interp(X,x,y)
plt.plot(x,y,marker = 'o')
```

```python
plt.plot(X,Y,marker = 'o')
```
